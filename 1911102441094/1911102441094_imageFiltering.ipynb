# filter kecerahaan

```python
img = PIL.Image.open('sample_data/12-min.jpg') 
converter = ImageEnhance.Brightness(img) # variable converter dimana didalam variabel tersebut berisikan fungsi brightness
img2 = converter.enhance(0.5) 
img3 = converter.enhance(2)
```

### Menampilkan gambar asli, setengah terang dan terang

```python
plt.gcf().set_size_inches(15,16) # ukuran foto yang di tampilkan 15 panjang 16 tinggi
plt.subplot(131),plt.imshow(img),plt.title('Original')  
plt.xticks([]), plt.yticks([]) 
plt .subplot(132),plt.imshow(img2),plt.title('Half Brightness') 
plt.xticks([]), plt.yticks([]) 
plt.subplot(133),plt.imshow(img3),plt.title('Double Brightness') 
plt.xticks([]), plt.yticks([]) 

plt.imshow(img)
plt.imshow(img2)
plt.imshow(img3)
```

### pada baris 2, 4 dan 6 yang di dalam kurung contoh plt.subplot(131) berarti 1 merupakan jumlah kolom yang di tampilkan, 3 merupakan jumlah gambar yang akan di tampilkan dalam kolom itu berjumlah 3 gambar, digit ke tiga dalam angka tersebut berarti menampilkan urutan gambar dalam kolom.


```python

img = cv2.imread('sample_data/12-min.jpg',0)
img = cv2.medianBlur(img,5)

ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,11,2)
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,2)
titles = ['Original Image', 'Global Thresholding (v = 127)',
            'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [img, th1, th2, th3]
for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
